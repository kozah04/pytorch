```python
#2 random tensor woth shape(7,7)
import torch
tensor = torch.rand(7,7)
tensor
```

```python
#3 matrix mul on tensor with another tensor (1,7)
# creating tensor2
tensor2 = torch.rand(1,7)
tensor2
```

```python
tensor_mul = tensor.matmul(tensor2.T)
tensor_mul
```

```python
#4 set random seed to 0 and do exercises 2 and 3 again

# setting rand seed
torch.manual_seed(0)
tensor = torch.rand(7,7)

tensor2 = torch.rand(1,7)

tensor_mul = tensor.matmul(tensor2.T)
tensor_mul
```

```python
#5 set the GPU random seed to 1234
torch.cuda.manual_seed(1234)
```

```python
# 6 create rand tensors of shape (2,3) and send them both to GPU
torch.manual_seed(1234)
device = "cuda" if torch.cuda.is_available() else "cpu"
tensor1 = torch.rand(2,3).to(device)
tensor2 = torch.rand(2,3).to(device)

tensor1, tensor2
```

```python
# 7 perform matrix mul on the tensors you created in 6

tensor_mul = tensor1.matmul(tensor2.permute(1,0))
tensor_mul
```

```python
#8 find min and max of tensor_mul
tensor_mul.max(), tensor_mul.min()
```

```python
#9 find index of min and max
tensor_mul.argmax(), tensor_mul.argmin()
```
